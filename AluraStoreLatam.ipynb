### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
total_precio_tienda_1 = tienda['Precio'].sum()
total_precio_tienda_2 = tienda2['Precio'].sum()
total_precio_tienda_3 = tienda3['Precio'].sum()
total_precio_tienda_4 = tienda4['Precio'].sum()

print("Total precio tienda 1: ", total_precio_tienda_1)
print("Total precio tienda 2: ", total_precio_tienda_2)
print("Total precio tienda 3: ", total_precio_tienda_3)
print("Total precio tienda 4: ", total_precio_tienda_4)

```

```python
# Calcular ingresos totales
ingresos = {
    "Tienda 1": tienda["Precio"].sum(),
    "Tienda 2": tienda2["Precio"].sum(),
    "Tienda 3": tienda3["Precio"].sum(),
    "Tienda 4": tienda4["Precio"].sum()
}


print("="*40)
print("     REPORTE DE INGRESOS POR TIENDA     ")
print("="*40)
for tienda, ingreso in ingresos.items():
    print(f"{tienda:<10} ->  ${ingreso:,.2f}")
print("-"*40)
print(f"{'Total':<10} ->  ${sum(ingresos.values()):,.2f}")
print("="*40)
```

# 2. Ventas por categoría

```python
import pandas as pd
import matplotlib.pyplot as plt


url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

def contar_categorias(df, nombre_tienda):
    conteo = df['Categoría del Producto'].value_counts()
    print(f"\n{'='*45}")
    print(f" CATEGORÍAS MÁS VENDIDAS - {nombre_tienda.upper()} ")
    print(f"{'='*45}")
    for categoria, cantidad in conteo.items():
        print(f"{categoria:<20} -> {cantidad} ventas")
    print(f"{'-'*45}")


contar_categorias(tienda, "Tienda 1")
contar_categorias(tienda2, "Tienda 2")
contar_categorias(tienda3, "Tienda 3")
contar_categorias(tienda4, "Tienda 4")
```

En este paso, debes calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.

```python
#En este paso, debes calcular el costo de envío promedio para cada tienda.
# El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.

import pandas as pd



url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

def calcular_envio_promedio(df, nombre_tienda):
    df.columns = df.columns.str.strip()
    if 'Costo de envío' in df.columns:
        promedio = df['Costo de envío'].mean()
        print(f"{nombre_tienda:<10} ->  ${promedio:,.2f} MXN en promedio")
        return promedio
    else:
        print(f"[ERROR] No se encontró la columna 'Costo del Envío' en {nombre_tienda}")
        return None

print("="*50)
print("     COSTO PROMEDIO DE ENVÍO POR TIENDA     ")
print("="*50)
promedios_envio = {
    "Tienda 1": calcular_envio_promedio(tienda, "Tienda 1"),
    "Tienda 2": calcular_envio_promedio(tienda2, "Tienda 2"),
    "Tienda 3": calcular_envio_promedio(tienda3, "Tienda 3"),
    "Tienda 4": calcular_envio_promedio(tienda4, "Tienda 4")
}
print("="*50)
```

```python
import matplotlib.pyplot as plt


promedios_validos = {k: v for k, v in promedios_envio.items() if v is not None}

# Crear gráfico de pastel
plt.figure(figsize=(6, 6))
plt.pie(
    promedios_validos.values(),
    labels=promedios_validos.keys(),
    autopct='%1.1f%%',
    startangle=90,
    colors=['#FF9999', '#66B3FF', '#99FF99', '#FFCC99'],
    wedgeprops={'edgecolor': 'black'}
)
plt.title("Proporción del costo promedio de envío por tienda")
plt.axis('equal')
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

```python

def mostrar_productos(tienda, nombre_tienda):

    conteo_productos = tienda['Producto'].value_counts()


    top_productos = conteo_productos.head(5)


    bottom_productos = conteo_productos.tail(5)


    print(f"\n{'='*50}")
    print(f"     PRODUCTOS MÁS Y MENOS VENDIDOS - {nombre_tienda.upper()}     ")
    print(f"{'='*50}")


    print(f"\n--- PRODUCTOS MÁS VENDIDOS ---")
    for producto, cantidad in top_productos.items():
        print(f"{producto:<50} -> {cantidad} ventas")


    print(f"\n--- PRODUCTOS MENOS VENDIDOS ---")
    for producto, cantidad in bottom_productos.items():
        print(f"{producto:<50} -> {cantidad} ventas")

    print(f"{'-'*50}")


mostrar_productos(tienda, "Tienda 1")
mostrar_productos(tienda2, "Tienda 2")
mostrar_productos(tienda3, "Tienda 3")
mostrar_productos(tienda4, "Tienda 4")


```
